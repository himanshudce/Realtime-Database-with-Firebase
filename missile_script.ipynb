Import all libraries

```python
import requests
import json
import time
#sudo kill -9 $(sudo lsof -t -i:3001)
```

Functions for loading the data and counting rows in meanwhile

```python
# Postgres
def load_data_postgres(data):
    start = time.time()
    time_list = []
    prev_time = start
    dif = 1
    for line in data:
        response_insert = requests.post('http://localhost:3001/postgres/write/missile',json=line).text
        if (response_insert == "missile"):
            print("Missile detected at:", (time.time()-start))
        dt = time.time() - prev_time
        if dt > 1:
            # print or write to log here
            prev_time = time.time()   
            response_write = int(requests.get('http://localhost:3001/postgres/read/missile').text)
            time_list.append((dif*10,response_write))
            dif+=1
            print(response_write)
    end = time.time()
    print("postgres",end-start)
    return time_list

# Firebase
def load_data_firebase(data):
    start = time.time()
    time_list = []
    prev_time = start
    dif=1
    for line in data:
        response_insert = requests.post('http://localhost:3001/firebase/write/missile',json=line).text
        if (response_insert == "missile"):
            print("Missile detected at:", (time.time()-start))
        dt = time.time() - prev_time
        if dt > 10:
            # print or write to log here
            prev_time = time.time()   
            response_write = int(requests.get('http://localhost:3001/firebase/read/missile').text)
            time_list.append((dif*10,response_write))
            dif+=1
            print(response_write)
    end = time.time()
    print("firebase",end-start)
    return time_list
```

Select which data to load

```python
json_file_path= './data/trajectory_data_60k.json'
with open(json_file_path) as json_file:
    data = json.load(json_file)
```

Run process for Firebase

```python
time_list = load_data_firebase(data)
```
