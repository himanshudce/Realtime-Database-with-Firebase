```python
duration = []
x=[]
y=[]
for i in range(3000000):
    temp = time.time()
    y.append(i/1000)
    x.append(i/1000)
    duration.append(temp)
    time.sleep(0.0002)
```

```python
x=x[::-1]
```

```python
y=y[::-1]
```

```python
trg_data_final = list(zip(duration,x,y))
```

```python
trg_json = []
for row in trg_data_final:
    temp = {}
    temp['time'] = row[0]
    temp['x'] = row[1]
    temp['y'] = row[2]
    trg_json.append(temp)
```

```python
data_btc = pd.read_csv('data/gemini_BTCUSD_2018_1min.csv')
data_eth = pd.read_csv('data/gemini_ETHUSD_2018_1min.csv')
data_ltc = pd.read_csv('data/gemini_LTCUSD_2018_1min.csv')
```

```python
min_len = min(len(data_btc),len(data_eth),len(data_ltc))
min_len
```

```python
# importing the multiprocessing module
import multiprocessing
  
def print_cube(num):
    """
    function to print cube of given num
    """
    #print("Cube: {}".format(num * num * num))
    print(time.time())
    for i in range(10):
        print(i)
        print()
  

def print_square(num):
    """
    function to print square of given num
    """
    #print("Square: {}".format(num * num))
    print(time.time())
    for i in range(10,20):
        print(i)
        print()
        
  
if __name__ == "__main__":
    # creating processes
    p1 = multiprocessing.Process(target=print_square, args=(10, ))
    p2 = multiprocessing.Process(target=print_cube, args=(10, ))
  
    # starting process 1
    p1.start()
    # starting process 2
    p2.start()

  
    # both processes finished
    print("Done!")
```
